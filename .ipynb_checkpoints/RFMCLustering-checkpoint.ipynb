```python
# Import required libraries
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Load the data from an Excel file
data = pd.read_excel('Online Retail.xlsx')
data.head()
```

```python
# Check the information about the data
data.info()
```

```python
# Remove rows with missing CustomerID
data = data[data['CustomerID'].notnull()]
data.info()
```

```python
# Convert InvoiceDate to a datetime object
data['InvoiceDay'] = data['InvoiceDate'].apply(lambda x: dt.datetime(x.year, x.month, x.day))
data.head()
```

```python
# Calculate the reference date
pin_date = max(data['InvoiceDay']) + dt.timedelta(1)
pin_date
```

```python
# Calculate the total sum for each transaction
data['TotalSum'] = data['Quantity'] * data['UnitPrice']
data.head()
```

```python
# Perform RFM analysis
rfm = data.groupby('CustomerID').agg({
    'InvoiceDay': lambda x: (pin_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalSum': 'sum'
})
rfm
```

```python
# Rename the columns for RFM analysis
rfm.rename(columns={
    'InvoiceDay': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)
rfm
```

```python
# Assign labels to each RFM component
r_labels = range(4, 0, -1)
r_groups = pd.qcut(rfm['Recency'], q=4, labels=r_labels)
f_labels = range(1, 5)
f_groups = pd.qcut(rfm['Frequency'], q=4, labels=f_labels)
m_labels = range(1, 5)
m_groups = pd.qcut(rfm['Monetary'], q=4, labels=m_labels)
```

```python
# Add the labels to the RFM dataframe
rfm['R'] = r_groups.values
rfm['F'] = f_groups.values
rfm['M'] = m_groups.values
rfm
```

```python
# Prepare the data for clustering
X = rfm[['R', 'F', 'M']]
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=400, random_state=0, algorithm='elkan')
kmeans.fit(X)
```

```python
# Get the cluster labels
kmeans.labels_
```

```python
# Add the cluster labels to the RFM dataframe
rfm['kmeans_cluster'] = kmeans.labels_
rfm
```

```python
# Analyze the average values for Cluster 3
rfm[rfm['kmeans_cluster'] == 3].mean()
```

```python
# Analyze the average values for all clusters
rfm.mean()
```
